## Dummy DataSets

```python
import pandas as pd
import numpy as np
```

```python
ndatos=500
genero=['masculino','femenino']
clases=['baja','media','alta']
genero_data=[]
clases_data=[]

for i in range(ndatos):
    genero_data.append(np.random.choice(genero))
    clases_data.append(np.random.choice(clases))


dummy_dataset=pd.DataFrame(
{
    'Edad': 25+5*np.random.randn(ndatos).astype(int),
    'Genero': genero_data,
    'Clase Social': clases_data
})
```

```python
dummy_dataset.head(10)
```

```python
dummy_dataset.sort_values(['Edad','Genero']) #ordenar dataset
```

## Agrupacion de Datos 

```python
genero_group=dummy_dataset.groupby(['Genero'])
for names,groups in genero_group:
    print(names)
    print(groups)

```

```python
dos_grupos=dummy_dataset.groupby(['Genero','Clase Social'])
for names,groups in dos_grupos:
    print(names)
    print(groups)
```
